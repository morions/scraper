```python
from bs4 import BeautifulSoup
import requests
```

```python
#def find_propriedades():
#getting the html script

url = 'https://www.vivareal.com.br/venda/minas-gerais/uberlandia/'
html_text = requests.get(url).text
soup = BeautifulSoup(html_text, 'lxml')

propriedades = soup.find_all('article', class_ = 'property-card__container js-property-card')
cont= 0

#Searching for propriedades
for propriedade in propriedades:

    endereco = propriedade.find('span', class_ = 'property-card__address').text.strip()
    metros = propriedade.find('span', class_ = 'property-card__detail-value js-property-card-value property-card__detail-area js-property-card-detail-area').text.strip()
    preco = propriedade.find('div', class_ = 'property-card__price js-property-card-prices js-property-card__price-small').text.strip()

    #cleaning link
    link = propriedade.a['href']



    #isolating 'propriedade' type
    tipotxt = propriedade.find('span', class_ = 'property-card__title js-cardLink js-card-title').text
    tiposplit = tipotxt.split()
    tipo = tiposplit[0]

    #counting
    cont = cont + 1



    #print(propriedade)
    print(endereco)
    print(metros,'m^2')
    print(tipo)
    print(preco)
    print('https://www.vivareal.com.br'+link)
    print(' ')


print(cont)
```
